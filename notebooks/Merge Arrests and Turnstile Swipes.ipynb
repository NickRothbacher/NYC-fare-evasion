```python
import pandas as pd
import datetime
```

```python
arrests = pd.read_csv('../data/Arrests_Assigned_250m.csv', index_col = 0)
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
x = []
for year in years:
    x.append(pd.read_csv('../data/Turnstile/{}_turnstile_objectid_format.csv'.format(year)))
    
turnstile = pd.concat(x)
```

```python
turnstile = turnstile.drop('Unnamed: 0', axis=1)
```

```python
arrests['ARREST_DATE'] = pd.to_datetime(arrests['ARREST_DATE'], format = '%m/%d/%Y')
```

```python
arrests['ARREST_DATE'].sort_values()
```

```python
turnstile.datetime = pd.to_datetime(turnstile['datetime'], format = '%Y-%m-%d')
```

```python
turnstile.datetime.sort_values()
```

```python
turnstile = turnstile.rename({'datetime':'ARREST_DATE'}, axis=1)
```

```python
arrests_turns = arrests.merge(turnstile, how ='left', on = ['ARREST_DATE', 'objectid'])
```

```python
#next we'll delete null values from the dataset (some arrests occur before we have turnstile data)
print('Null entry diffs', arrests_turns.entries.isnull().sum())
print('Null exit diffs', arrests_turns.exits.isnull().sum())
arrests_turns.dropna(subset = ['entries', 'exits'], how='any', inplace=True)
print('Clean Data len:', len(arrests_turns))
```

```python
stations = arrests_turns.groupby('objectid').agg({'ARREST_KEY': 'count'}).rename({'ARREST_KEY': 'arrests'}, axis=1)
```

```python
stations= stations.merge(turnstile.groupby('objectid').sum()['entries'].reset_index(), how='left', on ='objectid')
```

```python
stations = stations.drop_duplicates()
```

```python
stations['normalized'] = stations['arrests']*100000/stations['entries']
```

```python
stations= stations.merge(arrests[['objectid', 'name', 'line']], how='left', on ='objectid')
```

```python
stations = stations.drop_duplicates()
```
