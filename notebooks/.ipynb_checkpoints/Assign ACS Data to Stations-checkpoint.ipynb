```python
import pickle
import pandas as pd
```

```python
with open('../data/Census Tracts within 600m by station.pkl', 'rb') as f:
    tracts_by_station = pickle.load(f)
```

```python
ACS_data = pd.read_csv('../data/ACS data/ACS 5 year estimates 2011-2015 NYC tracts.csv').set_index('Unnamed: 0')
```

```python
Stations_DFs = {}
for station, tracts in tracts_by_station.items():
    Stations_DFs[station] = ACS_data[ACS_data.index.isin(tracts)]
```

```python
Stations_totals = {}
for station, df in Stations_DFs.items():
    Stations_totals[station] = df.sum(axis=0)
```

```python
final = pd.DataFrame(Stations_totals).T
```

```python
final.index = final.index.rename('stationid')
```

```python
final.to_csv('../data/Station area ACS data.csv')
```
